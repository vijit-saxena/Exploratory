```python
from sklearn import tree
dtree = tree.DecisionTreeClassifier(criterion='gini', max_depth=10)
```

```python
import csv
total_data = []
train_data_features = []
train_data_labels = []
val_data_features = []
val_data_labels = []
with open("train_change1.01.csv") as tdata:
    r = csv.reader(tdata)
    for row in r:
        del row[0]
        total_data.append(row)
        '''if i <= 1851:
            train_data_labels.append(row[-1])
            del row[-1]
            train_data_features.append(row)
        else:
            val_data_labels.append(row[-1])
            del row[-1]
            val_data_features.append(row)
        i += 1
        '''
del total_data[0]

random.shuffle(total_data)
for i in range(len(total_data)):
    if i<1851:
        train_data_labels.append(total_data[i][-1])
        del total_data[i][-1]
        train_data_features.append(total_data[i])
    else:
        val_data_labels.append(total_data[i][-1])
        del total_data[i][-1]
        val_data_features.append(total_data[i])
train_data_features = np.array(train_data_features)
train_data_labels = np.array(train_data_labels)
val_data_features = np.array(val_data_features)
val_data_labels = np.array(val_data_labels)
#print(train_data_labels)
#del train_data_features[0]
#del train_data_labels[0]
#print(val_data_labels)
#print(len(val_data_features))
```

```python
dtree = dtree.fit(train_data_features, train_data_labels)
```

```python
import graphviz
g = tree.export_graphviz(dtree, out_file=None)
graph = graphviz.Source(g)
```

```python
graph
```

Now we calculate the accuracy over our validation data and check for any overfitting 

```python
output = dtree.predict(val_data_features)
#dtree.score(val_data_features, val_data_labels)
```

```python
count = 0
for i in range(len(output)):
    if output[i] == val_data_labels[i]:
        count += 1
accuracy = count*100/len(output)
accuracy
```

Now taking output for the test file given..

```python
test_data_features = []
test_output_Dtrees = []
with open("test_change1.01.csv") as test_data:
    r = csv.reader(test_data)
    for row in r:
        del row[0]
        test_data_features.append(row)
del test_data_features[0]
#print(test_data_features)
test_output_Dtrees = dtree.predict(test_data_features)
```

until now we have created a decision tree and obviously it needs prunning and its depth to be altered..
the mentioned taks are implemented below..

```python
x = []
y = []
for i in range(1,100):
    x.append(i)
    count = 0
    estimate = tree.DecisionTreeClassifier(max_depth=i)
    estimate = estimate.fit(train_data_features, train_data_labels)
    out = estimate.predict(val_data_features)
    for j in range(len(out)):
        if out[j] == val_data_labels[i]:
            count += 1
    accuracy = count*100/len(out)
    y.append(accuracy)
    #print(str(accuracy) + '%' + ' for depth =' + str(i))
```

```python
plt.plot(x,y)
plt.show()
```
