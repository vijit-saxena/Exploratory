```python
from sklearn import tree
dt = tree.DecisionTreeClassifier()
```

```python
import csv
import random
data = []
#file = open('/home/vijit/LocalDiskE/Analiticity/train.csv', 'rb')
with open('change.csv') as file:
    reader = csv.reader(file)
    for row in reader:
        #target.append(list(row[-1]))
        #del row[-1]
        del row[0]
        data.append(row)
del data[0]
data = np.array(data)




print(data)
```

```python
#del target[0]
random.shuffle(data)
X = data[:1500]
Y = data[1500:]
train_data = []
val_data = []
train_target = []
val_target = []
#print(len(Y))
for row in X:
    train_target.append(list(row[-1]))
    del row[-1]
for row in Y:
    val_target.append((row[-1]))
    del row[-1]
train_data = X
val_data = Y
```

```python
dt = dt.fit(train_data, train_target)
```

```python
import graphviz
dot_data = tree.export_graphviz(dt, out_file=None) 
graph = graphviz.Source(dot_data)
```

```python
#graph
```

```python
output = dt.predict(val_data)
```

```python
count = 0
for i in range(len(val_target)):
    if val_target[i] == output[i]:
        count += 1
count /= len(val_target)
print(count*100)
```
